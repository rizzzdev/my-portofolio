## Pertanyaan 3:

## Pertanyaan 4:
Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom `BMI`!

### Jawab:
Sebelum menambahkan kolom baru, kita perlu menghitung terlebih dahulu BMI dari setiap siswa, yang nantinya akan digunakan untuk mengisi kolom baru yang akan dibuat. Dengan menggunakan program yang telah dibuat di atas, BMI setiap siswa dapat dihitung dengan:

```python
bmi_list= []
for i in range(len(df)):
    berat_badan  = df['Berat Badan (kg)'][i]
    tinggi_badan = df['Tinggi Badan (cm)'][i] / 100
    bmi          = bmi_calc(berat_badan, tinggi_badan)
    bmi_list.append(bmi)
```

hasil perhitungan BMI setiap siswa disimpan dalam variable `bmi_list`. Variable tersebut akan digunakan untuk mengisi kolom `BMI` yang akan dibuat seperti pada di bawah ini:

```python
df['BMI'] = bmi_list
df
```

## Pertanyaan 5:
Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI!
- Normal: 18.5-24.9
- Overweight: 25.0-29.9
- Obesity: 30.0-34.9
- Underweight: < 18.5


### Jawab:
Berikut ini adalah program untuk menentukan (classifier) status BMI untuk setiap siswa:

```python
def status_bmi_cls(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi < 29.9:
        return 'Overweight'
    elif 30.0 <= bmi < 34.9:
        return 'Obesity'
    else:
        return 'Status Tidak Diketahui'
```

Jika program tersebut diterapkan pada dataframe untuk setiap siswa, maka akan menjadi seperti berikut ini:

```python
status_bmi_list = []
for i in range(len(df)):
    bmi = df['BMI'][i]
    status_bmi = status_bmi_cls(bmi)
    status_bmi_list.append(status_bmi)
```

Hasil classifier status BMI dari setiap siswa disimpan dalam variable `status_bmi_list`. Berikut ini adalah perintah untuk mengupdate dataframe (menambahkan kolom `Status BMI`) yang akan diisi dengan variable tersebut:

```python
df['Status BMI'] = status_bmi_list
df
```

# Soal 2:
Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:

Input
- Money Capital : 100000
- Investment Tenure (years) : 2

Output
- Year 1, your total investment : 125000.0
- Year 2, your total investment : 156250.0


## Jawab:
Disini akan dibuat suatu program yang akan kita nyatakan dalam function bernama `investasi()`:

```python
def investasi():
    money_capital     = int(input("Money Capital             = "))
    investment_tenure = int(input('Investment Tenure (years) = '))
    interest          = 0.25
    total_investment  = money_capital

    print(60*'-')
    if (money_capital < 0 or money_capital > 1000000000) or (investment_tenure < 0 or investment_tenure > 25):
        print('''

    Error: Nilai Money Capital dan/atau Investment Tenure tidak sesuai ketentuan

    Rentang Money Capital    --> (0-1000000000) Rupiah
    Rentang Invesment Tenure --> (0-25) tahun
        ''')

    else:
        for i in range(1, investment_tenure + 1):
            increment = interest * total_investment
            total_investment = total_investment + increment
            print(f'Year {i}, your total investment : {total_investment}')
    print(60*'-')
```

Untuk menjalankan program tersebut, kita hanya perlu memanggil function `investasi()`:

```python
investasi()
```
